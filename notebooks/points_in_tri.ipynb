```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def points_in_triangle(A, B, C, N, ):
    u = np.linspace(0, 1, num=int(N**0.5))
    v = np.linspace(0, 1, num=int(N**0.5))
    
    uu, vv = np.dstack(np.meshgrid(u, v)).reshape(-1, 2).T
    mask = vv <= 1 - uu
    
    return np.multiply.outer(A, uu[mask]) + np.multiply.outer(B, vv[mask]) + np.multiply.outer(C, 1 - uu[mask] - vv[mask])
```

```python
points = np.array([[-10, 0], [10, 0], [-10,20]])
plt.plot(*points_in_triangle(*points, 100), '.')
```
