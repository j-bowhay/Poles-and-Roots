```python
import numpy as np
import matplotlib.pyplot as plt
from mpltools import annotation
plt.style.use("/home/jakeb/development/poles-and-roots/notebooks/science.mplstyle")
```

```python
timing = np.load("timing3.npy")
iterations = np.arange(1, timing.size+1)
```

```python
fig, ax = plt.subplots()
y = np.cumsum(timing)
ax.loglog(y)
annotation.slope_marker((61, 29), (3, 1), size_frac=0.2, ax=ax, pad_frac=0.05)
ax.set_xlabel("Number of Iterations")
ax.set_ylabel("Runtime (s)")
ax.set_xlim([0, y.size])
plt.savefig("figures/AAA_runtime.pdf")
```
