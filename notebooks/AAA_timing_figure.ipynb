```python
import numpy as np
import matplotlib.pyplot as plt
from mpltools import annotation
plt.style.use(['science','ieee'])
plt.rcParams['font.size'] = 13
```

```python
timing = np.load("timing2.npy")
iterations = np.arange(1, timing.size+1)
```

```python
fig, ax = plt.subplots()
ax.loglog(np.cumsum(timing))
annotation.slope_marker((55, 30), (3, 1), size_frac=0.2, ax=ax)
ax.set_xlabel("Number of Iterations")
ax.set_ylabel("Runtime (s)")
plt.savefig("figures/AAA_runtime.pdf")
```
