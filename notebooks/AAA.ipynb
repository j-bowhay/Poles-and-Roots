# AAA

In this notebook we test AAA on a number of test problems.

## The Gamma Function

First plot the original function:

```python
fig, ax = plt.subplots()
im = phase_plot(scipy.special.gamma, ax, domain=[-10, 10, -10, 10])
fig.colorbar(im)
```

Now compare against AAA approximation:

```python
N = 500
Z = np.linspace(-5, 5, N)
F = scipy.special.gamma(Z)

r = AAA(F, Z)

fig, ax = plt.subplots()
im = phase_plot(r, ax, domain=[-10, 10, -10, 10])
fig.colorbar(im)
```
