```python
from sympy import *
init_printing()
```

```python
k, l, y, d = symbols("k lambda y d")
k1 = sqrt(k**2 - l**2)
k2 = sqrt(1 - l**2)
```

```python
N = k2*sin(k1*y)*cos(k2*d) + k1*cos(k1*y)*sin(k2*d)
```

```python
N
```

```python
from sympy.printing.numpy import NumPyPrinter

NumPyPrinter().doprint(N).replace("numpy", "np").replace("lambda_","z").replace("y","y_prime")
```
