```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def points_in_triangle(A, B, C, N):
    u = np.linspace(0, 1, num=N)
    v = np.linspace(0, 1, num=N)
    
    uu, vv = np.meshgrid(u, v)
    uu = uu.ravel()
    vv = vv.ravel()
    
    return np.multiply.outer(A, uu) + np.multiply.outer(B, vv) + np.multiply.outer(C, 1-uu-vv)
```

```python
points = np.array([[0, 0], [2, 0], [1,1]])
plt.plot(*points_in_triangle(*points, 3), '.')
```
