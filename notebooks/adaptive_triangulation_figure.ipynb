```python
import numpy as np
import matplotlib.pyplot as plt
from poles_roots.triangulation import adaptive_triangulation
from poles_roots.reference_problems import func5, func5_prime
from poles_roots.plotting import phase_plot
plt.style.use("/home/jakeb/development/poles-and-roots/notebooks/science.mplstyle")
```

```python
rng = np.random.default_rng(1)

tri, arg = adaptive_triangulation(func5, func5_prime, initial_points=[-5.1 - 11j, 11 - 11j, 11 + 11j, -5.1 + 11j],
                                arg_principal_threshold=3.1, rng=rng)
```

```python
fig, ax = plt.subplots()
phase_plot(func5, ax, domain=[-6, 12, -12, 12])
ax.triplot(tri.points[:, 0], tri.points[:, 1], tri.simplices)
ax.plot(tri.points[:,0], tri.points[:,1], 'bo', zorder=11)
for i, simplex in enumerate(tri.simplices):
    centre = tri.points[simplex, :].mean(axis=0)
    ax.text(
        centre[0],
        centre[1],
        f"{round(np.real(arg[i]))}",
        color="w",
        horizontalalignment='center',
        verticalalignment='center',
        zorder=12)
plt.savefig("figures/subdivision_3.pdf")
```
