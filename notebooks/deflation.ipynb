```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/home/jakeb/development/poles-and-roots/notebooks/science.mplstyle")
```

```python
x = np.linspace(-2, 2, num=int(1e6))
eps = 1e-5
fig, ax = plt.subplots()
ax.plot(x, 0*x, '--')
ax.plot(x, np.sin(np.pi*x), '-', label=r"$\sin(\pi x)$")
ax.plot(x, np.sin(np.pi*x)/(x-1), '-', label=r"$\frac{\sin(\pi x)}{(x-1)}$")
ax.plot(x, np.sin(np.pi*x)/(x-(1+eps)), ':', label=r"$\frac{\sin(\pi x)}{(x-(1+\epsilon))}$", linewidth=2)

ax.set_xlim([-2, 2])
ax.set_ylim([-5, 1.5])
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend(loc='lower left')
plt.savefig("figures/deflation.pdf")
```
